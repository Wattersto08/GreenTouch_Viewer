```python
filename = 'GoogleEarth_KML/Bodium_Vineyard.kml'
```

```python
import folium as f 
import kml_parser_functs as kml
import map_functs as mf 
```

```python


name, features = kml.parse_KML(filename)


print(name)
```

```python
# Render Map
#m = f.Map(location=[51.05531670367141, 0.5354980472325033], zoom_start=15)

m = mf.generatemap([51.00551148675169, 0.5432519630088373])
```

```python
def plot_features(features,MAP):
    for feature in features:

        if feature.Type == 'Point':
            f.Marker(feature.midpoint, popup=feature.name, color=feature.color).add_to(MAP)
        elif feature.Type == 'Linesting':
            f.PolyLine(feature.coords, color = feature.color).add_to(MAP)
        else:
            f.Polygon(feature.coords, color = feature.color).add_to(m)
            f.Marker(feature.midpoint, color = feature.color, popup = feature.name).add_to(MAP)
```

```python
plot_features(features,m)
m
```
